```python
import pandas as pd
import os
import glob

# csv 파일 정리해서 합치기
folder_path = './data/지역연도별방문자추이/'

csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

dataframes = []
for file in csv_files:
    try:
        df = pd.read_csv(file, encoding='euc-kr')

        #index 열 생성.
        df['index'] = df['축제명'] + df['개최년도'].astype(str)
        dataframes.append(df)
    except Exception as e:
        print(f"Error reading {file}: {e}")

combined_df = pd.concat(dataframes, ignore_index=True)

print(combined_df)


#파일 저장 장소 지정. 이름도 지정.              # 바로 아랫줄 맨 앞 # 지우고 사용
#combined_df.to_csv('./combined_data.csv', index=False, encoding='euc-kr')
```

```python
# 엑셀 파일들 정리해서 합치기

import pandas as pd
import os
import glob

# 엑셀 파일이 저장된 폴더 경로 지정
folder_path = './회계정보전처리/'

excel_files = glob.glob(os.path.join(folder_path, '*.xlsx'))

dataframes = []
for file in excel_files:
    try:
        df = pd.read_excel(file)

        # 필요 없는 column전처리. 시작월, 종료월 column 생성, 'index' 열 생성, 자치단체코드, 개최시작일자, 개최종료일자, 사업내용, No, 열 삭제.
        df['시작월'] = df['개최시작일자'].astype(str).str[4:6].astype(int)
        df['종료월'] = df['개최종료일자'].astype(str).str[4:6].astype(int)
        df['index'] = df['행사축제명'] + df['회계연도'].astype(str)
        df = df.drop(['자치단체코드', '개최시작일자', '개최종료일자', '사업내용', 'No'], axis = 1)
        dataframes.append(df)
    except Exception as e:
        print(f"Error reading {file}: {e}")

if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    print(combined_df)
else:
    print("엑셀 내용 확인할 것.")



#파일 저장 장소 지정. 이름도 지정.              # 바로 아랫줄 맨 앞 # 지우고 사용
#combined_df.to_csv('./엑셀묶음파일.csv', index=False)

```

```python
import pandas as pd
import os
import glob

# 폴더 경로 지정
folder_path = './데이터랩/'

# 이름에 규칙 있는 파일만 불러와서 합치기
pattern = os.path.join(folder_path, '*_목적지 검색순위.csv')
csv_files = glob.glob(pattern)



dataframes = []
for file in csv_files:
    try:
        file_name = os.path.basename(file)       # 파일 이름만 추출
        festival_name = file_name.split('_')[1]  # '강경젓갈축제' 부분 추출

        df = pd.read_csv(file, encoding='euc-kr')
        
        # 축제명 열 추가
        df['축제명'] = festival_name
        df['도'] = df['도로명주소'].str.split(' ').str[0]
        df['시군구'] = df['도로명주소'].str.extract(r'(\S+시|\S+군|\S+구)')  # '~~시', '~~군', '~~구'
        df = df.drop(['목적지명', '카테고리', '구분', '순위', '도로명주소'], axis = 1)
        
        dataframes.append(df)
    except Exception as e:
        print(f"이 파일 확인 {file}: {e}")


if dataframes: 
    combined_df = pd.concat(dataframes, ignore_index=True)
    print(combined_df)

    #파일 저장 장소 지정. 이름도 지정.              # 바로 아랫줄 맨 앞 # 지우고 사용
    combined_df.to_csv('./combined_data.csv', index=False, encoding='euc-kr')
else:
    print("파일 위치나 코드 확인할 것.")


```

```python
import pandas as pd
import os
import glob

# 폴더 경로 지정
folder_path = './data/데이터랩/'

# 이름에 규칙 있는 파일만 불러와서 합치기
pattern = os.path.join(folder_path, '*_연령별 내국인 방문자.csv')
csv_files = glob.glob(pattern)


dataframes = []
for file in csv_files:
    try:
        file_name = os.path.basename(file)       # 파일 이름만 추출
        festival_name = file_name.split('_')[1]  # '축제이름' 부분 추출

        df = pd.read_csv(file, encoding='euc-kr')
        
        # 축제명 열 추가
        df['축제명'] = festival_name
        dataframes.append(df)
    except Exception as e:
        print(f"이 파일 확인 {file}: {e}")


if dataframes: 
    combined_df = pd.concat(dataframes, ignore_index=True)
    print(combined_df)

    #파일 저장 장소 지정. 이름도 지정.              # 바로 아랫줄 맨 앞 # 지우고 사용
    combined_df.to_csv('./data/성별,연령별데이터.csv', index=False, encoding='euc-kr')
else:
    print("파일 위치나 코드 확인할 것.")


```
