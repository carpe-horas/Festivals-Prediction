# 적합도 검정 2회 실시: 분포가 ~일것이라 가정한 그룹의비율 vs 관측데이터비율 테스트

```python
# 초기가정

# 귀무가설: 외지인과 현지인의 비율은 5:5 일것으로 가정한다(균등분포)
# 대립가설: 외지인과 현지인의 비율은 5:5가 아닐것으로 가정한다(비균등분포)

import numpy as np
import pandas as pd
from scipy.stats import chisquare

# 데이터 로드
df = pd.read_csv('Merged_TMAP_SearchRank_Rev.csv', encoding='utf-8')

# '구분' 열에서 '외지인'과 '현지인'만 필터링
filtered_df = df[df['구분'].isin(['외지인', '현지인'])]

# 관찰값 계산
observed_counts = filtered_df['구분'].value_counts()
observed = observed_counts.values  # 관찰값

# 기대값 계산 (균등 분포: 5:5 비율)
total = sum(observed)
expected = [total * 0.5, total * 0.5]  # 50% 비율로 기대값 계산

# 카이제곱 적합도 검정 수행
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# 결과 출력
print("Chi-squared statistic:", chi2_stat)
print("p-value:", p_value)

# 관찰값 및 기대값 확인
print("\nObserved counts:")
print(observed_counts)
print("\nExpected counts (equal distribution 50:50):")
print(expected)

# 귀무가설 기각 여부 판단
alpha = 0.05  # 유의수준 설정
if p_value <= alpha:
    print("\n귀무가설(기각): p_value가 유의수준으로 설정한 0.05보다 유의미하게 작기때문에 현지인과 외지인의 비율은 50:50일 것이라고 가정한다.")
else:
    print("\n귀무가설(선택): p_value가 유의수준으로 설정한 0.05보다 유의미하게 크기때문에 현지인과 외지인의 비율은 50:50가 아닐것으로 가정한다.")
```

```python
# 수정가정(시각화결과바탕,최적화(Δ｜예측값-실제값｜:최소값))

# 귀무가설: 외지인과 현지인의 비율은 외지인 0.5736 : 현지인 0.4264 일것으로 가정한다
# 대립가설: 외지인과 현지인의 비율은 외지인 외지인 0.5736 : 현지인 0.4264 아닐것으로 가정한다

import numpy as np
import pandas as pd
from scipy.stats import chisquare

# 데이터 로드
df = pd.read_csv('Merged_TMAP_SearchRank_Rev.csv', encoding='utf-8')

# '구분' 열에서 '외지인'과 '현지인'만 필터링
filtered_df = df[df['구분'].isin(['외지인', '현지인'])]

# 관찰값 계산
observed_counts = filtered_df['구분'].value_counts()
observed = observed_counts.values  # 관찰값

# 기대값 계산 (외지인 57.36%, 현지인 42.64% 비율)
total = sum(observed)
expected = [total * 0.5736, total * 0.4264]  # 57.36%와 42.64% 비율로 기대값 계산

# 카이제곱 적합도 검정 수행
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# 결과 출력
print("Chi-squared statistic:", chi2_stat)
print("p-value:", p_value)

# 관찰값 및 기대값 확인
print("\nObserved counts:")
print(observed_counts)
print("\nExpected counts (57.36%:42.64%):")
print(expected)

# 귀무가설 기각 여부 판단
alpha = 0.05  # 유의수준 설정
if p_value <= alpha:
    print("\n대립가설(기각): p_value가 유의수준으로 설정한 0.05보다 유의미하게 작기때문에 외지인과 현지인의 비율은 57.36:42.64가 아니라고 할 수 있습니다.")
else:
    print("\n귀무가설(선택): p_value가 유의수준으로 설정한 0.05보다 유의미하게 크기때문에 외지인과 현지인의 비율은 57.36:42.64라고 할 수 있습니다.")
```

# 독립성 검정 1회 실시: 행과(카테고리) 열이(검색순위) 서로 영향을 준다
# H₀귀무가설: 카테고리와 검색순위는 관련성이 없다, 독립이다(기각)
# Η₁대립가설: 카테고리와 검색순위는 관련성이 있다, 독립이 아니다(선택)

```python
from scipy.stats import chi2_contingency

# '구분' 열에서 '외지인'과 '현지인'만 필터링
filtered_df = df[df['구분'].isin(['외지인', '현지인'])]

# 교차표 생성
contingency_table = pd.crosstab(filtered_df['카테고리'], filtered_df['순위'])

# 카이제곱 독립성 검정 수행
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# 결과 출력
print(f"Chi-squared statistic: {chi2_stat}")
print(f"p-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print(f"Expected counts:\n{expected}")

# 귀무가설을 기각할지 결정
alpha = 0.05  # 유의수준 0.05 설정

if p_value < alpha:
    print("p_value값이 α값으로 설정한 0.05보다 유의미하게 작기때문에, 귀무가설을 기각합니다. 카테고리 변수와 순위 변수는 독립이 아닙니다.")
else:
    print("p_value값이 α값으로 설정한 0.05보다 유의미하게 크기때문에, 귀무가설을 선택합니다. 카테고리 변수와 순위 변수는 독립 입니다.")
```

# Z검정

```python
import pandas as pd

# 데이터 로드
df = pd.read_csv('./Merged_TMAP_SearchRank_Rev.csv', encoding='utf-8')

# '구분' 열의 유니크 값 확인
print(df['구분'].unique())

# '구분' 열에서 '미분류_현외' 값의 행만 선택
unclassified_count = df[df['구분'] == '미분류_현외'].shape[0]

# 비율에 맞게 '현지인'과 '외지인'으로 분할
local_count = int(unclassified_count * 0.4264)  # 현지인 비율
external_count = unclassified_count - local_count  # 외지인 비율

# '미분류_현외' 값을 '현지인'과 '외지인'으로 변경
df.loc[df['구분'] == '미분류_현외', '구분'] = (
    ['현지인'] * local_count + ['외지인'] * external_count
)

# '구분' 열의 값을 다시 확인
print(df['구분'].value_counts())

# 데이터 저장 (원본에 반영 후 저장)
df.to_csv('./Updated_TMAP_SearchRank_Rev.csv', index=False, encoding='utf-8')

# '구분' 열의 수정된 값을 확인
print(df.head())
```

```python
# 샘플집단의 비율이 모집단의 비율과 동등한지 Z검정

import pandas as pd
import numpy as np
from scipy import stats
%precision 4
pd.set_option("display.precision",4)

# 데이터 로드
df = pd.read_csv('./Updated_TMAP_SearchRank_Rev.csv', encoding='utf-8')

# 모집단 비율 (현지인 42.64%, 외지인 57.36%)
population_local_ratio = 0.4264
population_external_ratio = 0.5736

# 샘플링을 위한 설정
sample_size = 30  # 각 샘플의 크기

# 샘플링: 샘플을 추출하고 '구분' 열에서 현지인과 외지인의 비율 계산
sample = df['구분'].sample(n=sample_size, replace=True)  # 랜덤 샘플링
sample_local_count = sample.value_counts(normalize=True).get('현지인', 0)
sample_external_count = sample.value_counts(normalize=True).get('외지인', 0)

# 샘플 비율
sample_local_ratio = sample_local_count
sample_external_ratio = sample_external_count

# z-검정을 위한 계산
n = sample_size  # 샘플 크기
p_0 = population_local_ratio  # 모집단의 현지인 비율

# 샘플에서의 현지인 비율을 이용하여 z-검정
z_local = (sample_local_ratio - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)

# p-value 계산 (양측 검정)
p_value_local = 2 * (1 - stats.norm.cdf(abs(z_local)))

# 결과 출력
print(f"Sample Local Ratio: {sample_local_ratio:.4f}")
print(f"Sample External Ratio: {sample_external_ratio:.4f}")
print(f"Z-test for Local: Z = {z_local:.4f}, p-value = {p_value_local:.4f}")

# p-value가 0.05보다 크면 모집단 비율과 차이가 없다고 볼 수 있음
if p_value_local > 0.05:
    print("샘플집단의 local ratio는 모집단의 local ratio와 충분히 다르지 않다")
else:
    print("샘플집단의 local ratio는 모집단의 local ratio와 충분히 다르다")
```
